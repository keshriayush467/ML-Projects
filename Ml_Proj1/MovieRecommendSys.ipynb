```python
movies = movies[['movie_id','title', 'overview', 'genres', 'keywords', 'cast','crew']]
```

```python
movies.dropna(inplace =True)
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])

    return L
```

```python
movies["genres"]=movies['genres'].apply(convert)
```

```python
movies["keywords"]=movies['keywords'].apply(convert)
```

```python
# We need Only top 3 cast(characters for each movie)
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count +=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch)
```

```python
movies.head(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python

# filled the blank space 
# eg: Sam Worthington as SamWorthington
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
new_mov = movies[['movie_id', 'title', 'tags']] 
```

```python
new_mov.head()
```

```python
# Convert The list dtype in tag colm to string
new_mov['tags']= new_mov['tags'].apply(lambda x:" ".join(x))
```

```python
new_mov['tags'] = new_mov['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
# This we are doing to make common word 
# eg ['dance', 'dancing ', 'danced'] to a common word for all ['danc'] this process is steming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# defining the function
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_mov['tags']= new_mov['tags'].apply(stem)
```

```python
# Vectorizzation Convert the string in tags colm to a vectors of numbers which show repetion of similar or repeated word
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer( max_features= 5000,stop_words='english')
vectors = cv.fit_transform(new_mov['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# Recommend function
def recommend(movies):
    movie_index = new_mov[new_mov['title'] == movies ].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)), reverse= True, key= lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_mov.iloc[i[0]].title)
       
    
```

```python
recommend('Batman Begins')
```
